## aggregate blocks in one day as a session, and normalized by session

```python

in_context_LoRAtrialnum_MeanFr_normalized_sessions=[]
session_length = 2 # every 2 blocks are aggregated as a session
i=1
for context_order,context_angle,in_context_LoRAtrialnum_MeanFr_msblock in zip(context_orders,context_angles,in_context_LoRAtrialnum_MeanFr_msblocks):
    in_context_LoRAtrialnum_MeanFr_msblock["context_order"]=context_order
    in_context_LoRAtrialnum_MeanFr_msblock["context_angle"]=context_angle
    if i%session_length!=0:
        temp = []
        temp.append(in_context_LoRAtrialnum_MeanFr_msblock)
    else:
        temp.append(in_context_LoRAtrialnum_MeanFr_msblock)
        temp_session = pd.concat(temp)
        columns2norm = temp_session.columns.drop(["in_context_trialnum","in_context_LoRs","context_order","context_angle"])
        # Normalize every two blocks (one session)
        temp_session.loc[:,columns2norm] = Normalize_df(temp_session.loc[:,columns2norm])
        temp_normalized = temp_session
        
        temp_normalized["session_num"] = str(int(i/session_length))
        in_context_LoRAtrialnum_MeanFr_normalized_sessions.append(temp_normalized)
    i = i+1
in_context_LoRAtrialnum_MeanFr_normalized_sessions = pd.concat(in_context_LoRAtrialnum_MeanFr_normalized_sessions)
in_context_LoRAtrialnum_MeanFr_normalized_sessions
    
```
