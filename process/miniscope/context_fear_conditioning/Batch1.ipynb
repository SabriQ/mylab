```python
all_behave_videos_tracks = glob.glob(r"V:\miniscope\7_Miniscope_CHS\201909_tracking\201909\*.h5")
all_behave_videos_tracks

behave_videos = glob.glob(r"V:\miniscope\7_Miniscope_CHS\raw_data\2019\*\*\191126\*\[0-9]*[0-9].AVI")
behave_videos

behave_tracks=[]
for behave_video in behave_videos:
    key = str(re.findall('(\d{13}).AVI',behave_video)[0])
    [behave_tracks.append(i) for i in all_behave_videos_tracks if key in i]
behave_tracks

for behave_video,behave_track in zip(behave_videos,behave_tracks):
    dirname_video = os.path.dirname(behave_video)
    basename_track = os.path.basename(behave_track)
    new_behave_track_name = os.path.join(dirname_video,basename_track)
    shutil.copy(behave_track,new_behave_track_name)
    print(f"copied {new_behave_track_name}")

for behave_video in behave_videos:
    Video(behave_video).generate_ts_txt()
```

```python
def plot_trace(idxs,shocks=[5425,7227,9029],pre_duration=10,post_duration=12):
    plt.figure(figsize=(15,10))
    for j,shock in enumerate(shocks,1):
        plt.subplot(1,3,j)
        start_time=ts[shock]-pre_duration# pre_shock period start frame_num
        shock_ontime =ts[shock] # shock on time
        shock_offtime=ts[shock]+2
        end_time=ts[shock]+post_duration # post_shock period end frame_num
        
        start_frame=find_close_fast2(result["ms_ts"]/1000,start_time)
        end_frame = find_close_fast2(result["ms_ts"]/1000,end_time)
        x = (result["ms_ts"]/1000)[start_frame:end_frame]

        for i,idx in enumerate(idxs):
            y = temp[idx][start_frame:end_frame]+i*50
            plt.plot(x,y,linewidth=0.5,color="black")
        plt.axvspan(xmin=shock_ontime,xmax=shock_offtime,alpha=0.4,facecolor="blue")
        plt.yticks([])
        plt.xlabel("Time/s")
        plt.title("shock %s"%j)
        if j==1:
            plt.ylabel("Trace")
plot_trace(result["idx_accepted"])
```
